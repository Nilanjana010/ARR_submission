```python
import matplotlib.pyplot as plt
import re

# Create the directory if it doesn't exist
output_dir = "not_nuclear_plots"
os.makedirs(output_dir, exist_ok=True)

# Extract x-axis labels
x_labels = [list(d.keys())[0] for d in dt]
print("x_labels = ", x_labels)

# Define colors for each bar (10 different colors)
colors = ['skyblue', 'orange', 'green', 'blue', 'purple', 'brown', 'pink', 'yellow', 'cyan', 'magenta']

# Extract legend labels
legend_labels = [re.search(r'mul_adv_(.*?)_adversarial_texts.out_gr', label).group(1) for label in x_labels]
print("legend_labels = ", legend_labels)

# Keys to plot
keys_to_plot = [
    'crime_gr_count', 'crime_gr4_count', 'crime_k_5',
    'horror_gr_count', 'horror_gr4_count', 'horror_k_5',
    'war_gr_count', 'war_gr4_count', 'war_k_5']

for key in keys_to_plot:
    # Extract y-axis values for the current key
    y_values = [list(d.values())[0].get(key, 0) for d in dt]
    print("y_values = ", y_values)
    max_y = max([list(d.values())[0]['i_crime_count'] for d in dt])  # i_crime_count for y-axis limit
    print("max_y = ", max_y)
    
    # Plotting
    fig, ax = plt.subplots(figsize=(10, 6), facecolor='white')
    ax.set_facecolor('white')
    bars = ax.bar(x_labels, y_values, color=colors[:len(x_labels)], edgecolor='black')


    # Add legend for each color
    for bar, label in zip(bars, legend_labels):
        bar.set_label(label)
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(bar.get_height()), ha='center', va='bottom')
    
    if key == 'crime_gr_count':
         key = 'crime_gpt_4o_mini_5_count' 
    elif key == 'crime_gr4_count':
         key = 'crime_gpt_4o_mini_4_count'
    elif key == 'horror_gr_count':
         key = 'horror_gpt_4o_mini_5_count'
    elif key == 'horror_gr4_count':
         key = 'horror_gpt_4o_mini_4_count'
    elif key == 'war_gr_count':
         key = 'war_gpt_4o_mini_5_count'
    elif key == 'war_gr4_count':
         key = 'war_gpt_4o_mini_4_count'
    
    plt.axhline(y=max_y, color='blue', linestyle='--', label=f'Max i_crime_count = {max_y}')
    plt.xticks([])  # Remove x-axis labels
    plt.title(f"{key} by Model", fontsize=14)
    plt.xlabel("Models", fontsize=12)
    plt.ylabel(key, fontsize=12)
    plt.ylim(0, max_y + 5)  # Slightly higher than max for better visibility
    plt.legend(loc='upper right')  # Move legend to top right
    plt.tight_layout()

    # Save the plot to a file
    output_path = os.path.join(output_dir, f"{key}.png")
    plt.savefig(output_path)
    plt.show()

    plt.close()  # Close the figure to free up memory
```
